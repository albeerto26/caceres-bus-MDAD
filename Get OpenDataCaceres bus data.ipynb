```python
import pandas as pd
```

```python
df_routes = pd.read_csv('routes.txt')
del df_routes['route_url']
del df_routes['route_type']
del df_routes['agency_id']
df_routes
```

```python
df_trips = pd.read_csv('trips.txt')
# del df_trips['trip_id']
del df_trips['direction_id']
df_trips.head()
```

```python
route_id_dict = dict()
route_short_name_dict = dict()
route_long_name_dict = dict()
trip_id_dict = dict()
shape_id_dict = dict()
new_df_index = 0
for index_route, row_route in df_routes.iterrows():
    for index_trip, row_trip in df_trips.iterrows():
        if row_route['route_id'] == row_trip['route_id']:
            route_id_dict[new_df_index] = row_route['route_id']
            route_short_name_dict[new_df_index] = row_route['route_short_name']
            route_long_name_dict[new_df_index] = row_route['route_long_name']
            trip_id_dict[new_df_index] = row_trip['trip_id']
            shape_id_dict[new_df_index] = row_trip['shape_id']
            new_df_index += 1
            df_trips.drop(df_trips.index[index_trip])
            
route_id_serie = pd.Series(route_id_dict)
route_short_name_serie = pd.Series(route_short_name_dict)
route_long_name_serie = pd.Series(route_long_name_dict)
trip_id_serie = pd.Series(trip_id_dict)
shape_id_serie = pd.Series(shape_id_dict)
```

```python
df = pd.DataFrame()
df['route_id'] = route_id_serie
df['route_short_name'] = route_short_name_serie
df['route_long_name'] = route_long_name_serie
df['trip_id'] = trip_id_serie
df['shape_id'] = shape_id_serie
df.head()
```

```python
df_stops = pd.read_csv('stops.txt', names=['_key', 'name', 'lat', 'lon', 'url', 'wheelchair'], header=0)
df_stops.head()
```

```python
del df_stops['url']
del df_stops['wheelchair']
df_stops.head()
```

```python
df_stop_times = pd.read_csv('stop_times.txt')
df_stop_times.head()
df_stop_times.shape
```

```python
route_id_dict = dict()
route_short_name_dict = dict()
route_long_name_dict = dict()
trip_id_dict = dict()
shape_id_dict = dict()
stop_id_dict = dict()
stop_sequence_dict = dict()
stop_name_dict = dict()
lat_dict = dict()
lng_dict = dict()

new_df_index = 0

for index_df, row_df in df.iterrows():
    for index_stops_times, row_stops_times in df_stop_times.iterrows():
        if row_df['trip_id'] == row_stops_times['trip_id']:
            route_id_dict[new_df_index] = row_df['route_id']
            route_short_name_dict[new_df_index] = row_df['route_short_name']
            route_long_name_dict[new_df_index] = row_df['route_long_name']
            trip_id_dict[new_df_index] = row_df['trip_id']
            shape_id_dict[new_df_index] = row_df['shape_id']
            stop_id_dict[new_df_index] = row_stops_times['stop_id']
            stop_sequence_dict[new_df_index] = row_stops_times['stop_sequence']
            for index_stops, row_stops in df_stops.iterrows():
                if row_stops_times['stop_id'] == row_stops['_key']:
                    stop_name_dict[new_df_index] = row_stops['name']
                    lat_dict[new_df_index] = row_stops['lat']
                    lng_dict[new_df_index] = row_stops['lon']
                    break
            
            new_df_index += 1
            df_stop_times.drop(df_stop_times.index[index_stops_times])
            
route_id_serie = pd.Series(route_id_dict)
route_short_name_serie = pd.Series(route_short_name_dict)
route_long_name_serie = pd.Series(route_long_name_dict)
trip_id_serie = pd.Series(trip_id_dict)
shape_id_serie = pd.Series(shape_id_dict)
stop_id_serie = pd.Series(stop_id_dict)
stop_sequence_serie = pd.Series(stop_sequence_dict)
stop_name_serie = pd.Series(stop_name_dict)
lat_serie = pd.Series(lat_dict)
lng_serie = pd.Series(lng_dict)
```

```python
df = pd.DataFrame()
df['route_id'] = route_id_serie
df['route_short_name'] = route_short_name_serie
df['route_long_name'] = route_long_name_serie
df['trip_id'] = trip_id_serie
df['shape_id'] = shape_id_serie
df['stop_id'] = stop_id_serie
df['stop_sequence'] = stop_sequence_serie
df['stop_name'] = stop_name_serie
df['lat'] = lat_serie
df['lng'] = lng_serie
df.head()
```

```python
from_dict = dict()
to_dict = dict()
route_short_name_dict = dict()
route_long_name_dict = dict()
stop_from_name_dict = dict()
stop_to_name_dict = dict()
trip_id_dict = dict()

new_index = 0

for index, row in df.iterrows():
    df.drop(df.index[index])
    for index_, row_ in df.iterrows():
        if row['route_id'] == row_['route_id']:
            if row['stop_id'] != row_['stop_id']:
                from_dict[new_index] = 'stops/'+str(row['stop_id'])
                to_dict[new_index] = 'stops/'+str(row_['stop_id'])
                route_short_name_dict[new_index] = row['route_short_name']
                route_long_name_dict[new_index] = row['route_long_name']
                stop_from_name_dict[new_index] = row['stop_name']
                stop_to_name_dict[new_index] = row_['stop_name']
                trip_id_dict[new_index] = row['trip_id']
                new_index+=1
```

```python
new_df = pd.DataFrame()
new_df['_from'] = pd.Series(from_dict)
new_df['_to'] = pd.Series(to_dict)
new_df['route_short_name'] = pd.Series(route_short_name_dict)
new_df['route_long_name'] = pd.Series(route_long_name_dict)
new_df['stop_from_name'] = pd.Series(stop_from_name_dict)
new_df['stop_to_name'] = pd.Series(stop_to_name_dict)
new_df['trip_id'] = pd.Series(trip_id_dict)
new_df.head()
```

```python
df_frequencies = pd.read_csv('frequencies.txt')
df_frequencies.head()
```

```python
start_time_dict = dict()
end_time_dict = dict()
headway_secs_dict = dict()

for index, row in new_df.iterrows():
    for index_, row_ in df_frequencies.iterrows():
        if row['trip_id'] == row_['trip_id']:
            start_time_dict[index] = row_['start_time']
            end_time_dict[index] = row_['end_time']
            headway_secs_dict[index] = row_['headway_secs']
            break

new_df['start_time'] = pd.Series(start_time_dict)
new_df['end_time'] = pd.Series(end_time_dict)
new_df['headway_secs'] = pd.Series(headway_secs_dict)

del new_df['trip_id']
```

```python
new_df.to_csv('trips.csv', index=False)
```
